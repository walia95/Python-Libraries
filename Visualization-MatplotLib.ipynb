```python
## For importing matplotlib:
import matplotlib.pyplot as plt
```

```python
# Since the graphs created with matplot lib may not fit in the Jupyter notebook, we need to bring them down to scale.
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(1,5,10)
y = x**2
```

```python
plt.plot(x,y)
plt.show() #We use plt.show() method for printing the 
```

### If we want to name the  X-axis and Y-axis we can use the following methods:

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Simple Graph')
plt.plot(x,y)
plt.show()
```

```python
## Now if we want to print plots on the the same canvas we can use the plt.subplot method

plt.subplot(2,2,1)  #The arguments in subplot method are number of rows and columns.
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.5,0.5]) #Left,Bottom,Width,Height
axes2 = fig.add_axes([0.2,0.9,0.9,0.4])
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
```

```python
# We can add subplots using this object oriented method as well.
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(1,5,10)
y = x ** 2
# we use fig,axes for using printing multiple plots in the same canvas.

fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)

# Since we accessed the locations of the figure via For loop, we can access the locations usig the addresses as well.
```

```python
#Here we are accessing the plots one-by-one by using the addresses for the plots by indexes.
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y,'y')
axes[0].set_title('Plot 1')
axes[1].plot(y,x,'*')
axes[1].set_title('Plot2')
plt.tight_layout()
```

### Fig Size and DPI (Pixels per Inch)
* Fig Size is the size of the figure in inches and DPI stands for Dots Per Inch.
* We don't need to set custom DPI while using jupyter notebook.



```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1,0.1,0.8,0.8]) ## When we add axes to the figure, we can use decimals as well as integers.
ax.plot(x,y)


plt.show()
```

```python
## We can use arguments such as DPI'S, FigSize with multiple plots at the same time as well.

fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))
axes[0].plot(x,y,'^')
axes[1].plot(y,x,'*')
plt.tight_layout()
```

## Legend
* Legends tell us about the details of the graph, such as the parameters used and such.
* What does the x-axis contain, what do the different colors signify etc.
* The details of the legend are mentioned in the plots and to print th elegend in the figure we need to call the method ax.legend(). 
* We can pass an argument called LOC in the ax.legend() method so that no data gets hidden in the figure.
* LOC has many values b/w 0 to 11. All of them cover various areas in the graph. The value 0 is the optimum value as it selects the best fit for the placement of the legend.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label = 'x squared')
ax.plot(y,x, label='y cubed')
ax.legend(loc=0)
```

# PLOT APPEARENCE
* Plot appearence is the part where we edit the plots's appearence such as colors.
* When we plot the graph on the canvas we can also use the color argument to use names such as 'blue','red' etc.
* If we want custom colors we can use RGB Hex codes as arguments in colors as well. 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'red')
plt.show()
```

### We can also change the width of the line traced in the plot as well, with the "linewidth/ lw" argument. The default is 1. But we can change it as per our requirements.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green',linewidth=5)
plt.show()
```

### The alpha argument allows us to control the transparency of the line.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green',lw=10,alpha=0.1)
plt.show()
```

### The Linestyle argument allows us to tweak the pattern of the line.  There are various arguments which are possible for the design of the line as well. Instead of linestyle we can ise "ls" as well.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green',lw=5,alpha=0.7,ls = '-.')
plt.show()
```

### Markers are the points which can appear on the graph where X and Y intersect. We have the option of editing the markers as well such as their face colors, their patterns, their edge widths, their sizes etc.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green',lw=3,marker='o',markersize='10',markerfacecolor='red',
        markeredgewidth= '3',markeredgecolor='black')
plt.show()
```

### We can also cotrol the plot range of the plot as well i.e we can give the range in which we can limit the data points as well.


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green',lw=10,ls='-.')
ax.set_xlim([0,1])   #Here the arguments are the upper bound and the lower bounds.
ax.set_ylim([0,2]) 
plt.show()
```
